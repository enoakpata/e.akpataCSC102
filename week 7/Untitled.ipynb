```python
import pandas as pd

def jt_ventures(name, department):
    
    name = input('What is your name: ')
    department = input('What is your department: ')
    
    df = pd.read_csv("jt-ventures.csv")
    if df[(df['FIRST NAME'] == name) & (df['DEPARTMENT'] == department)].shape[0]>0:
        print(df)
        
    else:
        print('you are not an employee')
        
jt_ventures('name','department')
```

```python
class car():

    # init method or constructor
    def __init__(self, model, color):
        self.model = model
        self.color = color

    def show(self):
        print("Model is", self.model )
        print("color is", self.color )


toyota = car("Toyota Corolla", "blue")
kia = car("Kia Cerato", "green")

toyota.show() # same output as car.show(toyota)
kia.show() # same output as car.show(kia)
```
