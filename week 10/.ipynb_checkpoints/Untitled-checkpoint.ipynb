```python
class formula:
    
    def trapezium(self):
        height = float(input('Enter height: '))
        base1 = float(input('Enter base1: '))
        base2 = float(input('Enter base2: '))
       
        area = height/2 * (base1 + base2)
        print(f"The area of the trapezium is {area}")
    
    def rhombus(self):
        diagonal1 = float(input('Enter diagonal1: '))
        diagonal2 = float(input('Enter diagonal2: '))
           
        area1 = 1/2 * diagonal1 * diagonal2
        print(f"The area of the rhombus is {area1}")

    def parallelogram(self):
        base = float(input('Enter your base: '))
        altitude = float(input('Enter your altitude: '))
        
        area2 = base * altitude
        print(f"The area of the parallelogram is {area2}")
    
    def cube(self):
        length = int(input('Enter length: '))
        
        area3 = 6 * length ** 2
        print(f"The area of the cube is {area3}")
            
    def cylinder(self):
        radius = int(input('Enter radius: '))
        height = int(input('Enter height: '))
        
        volume = 3.14 * radius ** 2 * height
        print(f"The volume of the cylinder is {volume}")
        
f = formula()

equation = int(input('Area of trapezium(1)\nArea of rhombus(2)\nArea of parallelogram(3)\nArea of cube(4)\nVolume of cylinder(5)\nCHOOSE AN EQUATION: '))

if equation == 1:
    f.trapezium()
                
elif equation == 2:
    f.rhombus()
                
elif equation == 3:
    f.parallelogram()
      
elif equation == 4:
    f.cube()
    
elif equation == 5:
    f.cylinder()

else:
    print('Invalid input')

       
```
