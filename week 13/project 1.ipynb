```python
import pandas as pd

name = []
matric_no = []
department = []
level = []


students = int(input('How many students are there: '))

for i in range(students):
    Name = input('What is your name: ')
    Matric_no = input('What is your matric number: ')
    Department = input('What department are you in: ')
    Level = int(input('What level are you in: '))
    
    name.append(Name)
    matric_no.append(Matric_no)
    department.append(Department)
    level.append(Level)
    
info = {'Name': name, 'Matric Number': matric_no, 'Department': department, 'Level': level}

df = pd.DataFrame(info)
   
df.to_csv('PAU SMIS.csv')
df
```
